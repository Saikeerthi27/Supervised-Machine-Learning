```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_regression
```

```python
X, y = make_regression(n_samples=1000, n_features=2, noise = 10)
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.25, random_state = 50)
```

```python
from sklearn.neighbors import KNeighborsRegressor
knr_model = KNeighborsRegressor(n_neighbors=5)
```

```python
knr_model = knr_model.fit(X_train, y_train)
```

```python
knr_model
```

```python
y_pred = knr_model.predict(X_test)
y_pred
```

```python
from sklearn.metrics import r2_score

print(r2_score(y_test, y_pred))
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
params = {
    'n_neighbors' : [3,4,5],
    'weights' : ['uniform', 'distance'],
    'algorithm': ['ball_tree', 'kd_tree', 'brute'],
    'p' : [1,2]

}

#  When p = 1, this is equivalent to using manhattan_distance (l1), and euclidean_distance (l2) for p = 2
```

```python
from sklearn.model_selection import StratifiedKFold
cv = StratifiedKFold()
```

```python
gscv = GridSearchCV(estimator = knr_model, param_grid = params, cv=5, scoring = 'r2')
```

```python
gscv_model  = gscv.fit(X_train, y_train)
gscv_model
```

```python
y_pred_res = gscv_model.predict(X_test)
```

```python
y_pred_res
```

```python
from sklearn.metrics import r2_score

print(r2_score(y_test, y_pred_res))
```
