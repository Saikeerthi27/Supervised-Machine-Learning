##### ColumnTransformer: 
It allows you to apply different transformations to different columns of your dataset. Its especially useful when you have both numeric and categorical features that need different preprocessing steps.



##### transformers: 
A list of tuples specifying what transformations to apply and to which columns.

#### Format: ('name', transformer, column(s))

##### remainder='passthrough': 
##### Keeps all columns that are not explicitly transformed by ColumnTransformer. This means other columns will remain unchanged in the final output.

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

ct = ColumnTransformer(transformers = [('onehot', OneHotEncoder(drop='first'),[3])], remainder = 'passthrough' )
```

```python
type(X_train)
```

```python
X_train = ct.fit_transform(X_train)
X_train
```

```python
X_test = ct.fit_transform(X_test)
X_test
```

```python
X_train
```

```python
from sklearn.svm import SVR
model_r = SVR(kernel="sigmoid")
```

```python
svr_model = model_r.fit(X_train,y_train)
```

```python
test = svr_model.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r2 = r2_score(y_test,test)
ad_r2 = 1 - (1-r2)*((len(X_train)-1)/(len(X_train)-X_train.shape[1]-1))
print(r2)
print(ad_r2)
```
