```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

X['car_name'] = le.fit_transform(X['car_name'])
X['car_name']
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import StandardScaler

one_hot_cols = ['seller_type', 'fuel_type', 'transmission_type']
num_cols = [features for features in X.columns if df[features].dtype != 'O']
# num_cols = X.select_dtypes(exlcude='O').columns

preprocess = ColumnTransformer(transformers = [
    ('onehot', OneHotEncoder(), one_hot_cols),
    ('StandardScaling', StandardScaler(), num_cols)
], remainder = 'passthrough')
```

```python
X = preprocess.fit_transform(X)
```

```python
X = pd.DataFrame(X)
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.2, random_state = 50)
```

```python
X_train
```

```python
X_test
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
def evaluate(actual, predicted):
    mae = mean_absolute_error(actual, predicted)
    mse = mean_squared_error(actual, predicted)
    r2 = r2_score(actual, predicted)
    return mae, mse, r2
```

```python
models = {
    "Lr" : LogisticRegression(),
    "SVR": SVR(),
    "DT": DecisionTreeRegressor(),
    "KNR": KNeighborsRegressor(),
    "RFR": RandomForestRegressor()
}



for i in range(len(list(models))):
    model = list(models.values())[i]
    model_name = list(models.keys())[i]
    model.fit(X_train, y_train)

    y_train_predict = model.predict(X_train)
    y_test_predict = model.predict(X_test)
    print(f"------------------ Report for {model_name} -----------------------")


    train_mae,train_mse,train_r2 = evaluate(y_train,y_train_predict)
    test_mae,test_mse,test_r2 = evaluate(y_test,y_test_predict)

    print(list(models.keys())[i])
    
    print("---------------------------------------------------")

    print("Model performance for Training set")
    print("MAE: {:.4f}" .format(train_mae))
    print("MSE: {:.4f}" .format(train_mse))
    print("R2score: {:.4f}" .format(train_r2))

    print(".................................................")

    print(".................................................")
    print('Model Performance for Test set')
    print("MAE: {:.4f}" .format(test_mae))
    print("MSE: {:.4f}" .format(test_mse))
    print("R2score: {:.4f}" .format(test_r2))
```

## HYperparamter tuning

```python
knn_params = {
    "n_neighbors":[5,7,9,10],
    "weights": ['uniform', 'distance'],
    "algorithm": ['auto', 'ball_tree', 'kd_tree', 'brute']
}


rf_params = {"max_depth": [5,8,15,10],
            "max_features": [5,7,"auto",8],
            "min_samples_split": [2,8,15,20],
            "n_estimators": [100,150,200,500,1000]


}
```

```python
random_cv = [

    ('knn', KNeighborsRegressor(), knn_params),
    ("RF", RandomForestRegressor(), rf_params)
]
```

```python
from sklearn.model_selection import RandomizedSearchCV

from sklearn.model_selection import StratifiedKFold
cv = StratifiedKFold()


model_params = {}
for name, model, params in random_cv:
    randomcv_model= RandomizedSearchCV(estimator = model, param_distributions = params,cv=5, scoring = 'r2')
    randomcv_model.fit(X_train,y_train)
    model_params[name] = randomcv_model.best_params_
    model_score = randomcv_model.best_score_
    print(name, model_score)

for params in model_params:
    print("----------------- Best params --------------------- ")
    print(model_params[params])
```
